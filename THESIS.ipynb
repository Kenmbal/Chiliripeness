```python
!pip install roboflow

from roboflow import Roboflow
from ultralytics import YOLO

rf = Roboflow(api_key="4UetNZJemq4mrCoAdS5C")
project = rf.workspace("calamansi-classifier").project("chili-ripeness-detection-8drv7-dbdqb")
version = project.version(1)
dataset = version.download("yolov8")
```

```python
!ls {dataset.location}/
```

```python
# Load YOLOv8 model (pretrained on COCO for better results)
model = YOLO("yolov8s.pt")  # you can also use yolov8s.pt (slower but more accurate)

# Train
model.train(
    data=dataset.location + "/data.yaml",  # path to data.yaml from Roboflow
    epochs=50,        # number of training cycles
    imgsz=640,        # image size
    batch=16          # batch size
)
```

```python
# Load trained model
model = YOLO("runs/detect/train/weights/best.pt")

# Run prediction on an image
results = model.predict(source="/content/CHILI-RIPENESS-DETECTION-1/test/images/117_jpg.rf.3adc6fa810289671a76cb1896abac32a.jpg", show=True)  # replace with your chili test image
```
